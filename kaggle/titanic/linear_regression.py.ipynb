```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv("train.csv")
data
```

```python
train = pd.DataFrame(data, columns=['Pclass', 'Sex', 'Age', 'Parch', 'Fare'])
train
```

```python
train['Sex'] == 'male'
train['Sex'] = np.where(train['Sex'] == 'male', 0, 1)
train['Sex']
```

```python
train['Age'].mean()
```

```python
train['Age'].fillna(train['Age'].mean(), inplace=True)
```

```python
train['PclassSquare'] = train['Pclass'] ** 2
train['AgeSquare'] = train['Age'] ** 2
train['ParchSquare'] = train['Parch'] ** 2
train['FareSqure'] = train['Fare'] ** 2
train
```

```python
Y = data['Survived']
```

```python
Y
```

```python
regr = linear_model.LinearRegression()
regr.fit(train, Y)
regr.score(train, Y)
```
